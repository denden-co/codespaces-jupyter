My First ML Project

Load data

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
df
```

Data preparation as X and Y

```python
y = df['logS']
y
```

```python
x = df.drop(['logS'], axis=1)
x
```

Data splitting

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=100)
```

```python
x_train.shape
```

```python
x_test.shape
```

Model building
Linear Regression

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
LinearRegression()
```

Training the model
Applying the model to make prediction

```python
y_lr_train = lr.predict(x_train)
y_lr_test_pred = lr.predict(x_test)
```

```python
print(y_lr_train)
```

```python
y_lr_test_pred
```

Evaluate model performance

```python
y_train
```

```python
y_lr_test_pred
```

Evaluate model

```python
from sklearn.metrics import mean_squared_error, r2_score
lr_train_mse = mean_squared_error(y_train, y_lr_train)
lr_train_r2 = r2_score(y_train, y_lr_train)
```

```python
lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE (Train): ', lr_train_mse)
print('LR R^2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R^2 (Test): ', lr_test_r2)
```

```python
lr_train_mse
```

```python
lr_test_mse
```

```python
lr_test_r2
```

```python
lr_train_r2
```

```python
lr_results = pd.DataFrame(['Linear Regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Train MSE', 'Train R2', 'Test MSE', 'Test R2']
```

```python
lr_results
```

Random Forest
Training the model 
Applying the model to make a prediction
Evaluate model performance

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_train, y_train)
```

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)
rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Train MSE', 'Train R2', 'Test MSE', 'Test R2']
rf_results
```

Model comparison

```python
df_results = pd.concat([lr_results, rf_results], axis=0)
df_results
```

```python
df_results.reset_index(drop=True)
```
